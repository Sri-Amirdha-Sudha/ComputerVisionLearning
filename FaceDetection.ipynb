### Face detection

```python
# create an object for cascade classifier
cascade_obj=cv2.CascadeClassifier("C:\\Users\\Priya\\Anaconda3\\lib\\site-packages\\cv2\\data\\haarcascade_frontalface_alt.xml")
```

```python
#reading the image
image=cv2.imread("people.jpg")
```

```python
# converting it into grayscale
grayImage=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
#searching the co-ordinates
face=cascade_obj.detectMultiScale(grayImage,scaleFactor=1.05,minNeighbors=5)
```

```python
print(type(face))
```

```python
print(face)
```

```python
for (x,y,w,h) in face:
    image=cv2.rectangle(image,(x,y),(x+w,y+h),(110, 197, 233),5)
```

```python
#resize=cv2.resize(image,(400,400))
cv2.imshow("gray Image",image)
cv2.waitKey(4000)
cv2.destroyAllWindows()
```

```python
import cv2
myVideo=cv2.VideoCapture(0)
a=1
while True:#infinite loop for capturing image
    a=a+1
    check,frame=myVideo.read()
    print(frame,check)
    grayScale=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    face=cascade_obj.detectMultiScale(grayScale,scaleFactor=1.05,minNeighbors=5)
    for (x,y,w,h) in face:
        frame=cv2.rectangle(frame,(x,y),(x+w,y+h),(110, 197, 233),5)
    cv2.imshow("My Video",frame)
    key=cv2.waitKey(1)
    if key==ord('a'):
        break
print(a)
myVideo.release()
cv2.destroyAllWindows()
```
